# Handling imbalanced dataset

# Undersampling

```python
#Undersampling is a technique to handle imbalanced dataset
normal=df[df['Class']==0]
fraud=df[df['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
#selects 473 samples randomly
normal_sample=normal.sample(n=473)
normal_sample.shape
```

```python
#concat normal and fraud samples
new_data=pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Logistic Regression

```python
#Implementing logistic Regression on undersampling dataset

from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,Y_train)
```

```python
#Make predictions

Y_pred=log.predict(X_test)
```

```python
#Model Evaluation

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print("Accuracy score is:",accuracy_score(Y_test,Y_pred))
print("Precision is:",precision_score(Y_test,Y_pred))
print("Recall is:",recall_score(Y_test,Y_pred))
print("F1 score is:",f1_score(Y_test,Y_pred))
```

# Decision Tree Classifier

```python
#Implementing Decision Tree Classifier on undersampling dataset

from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
#Make prediction

Y_pred1=dtc.predict(X_test)
```

```python
print("Accuracy score is:",accuracy_score(Y_test,Y_pred1))
print("Precision is:",precision_score(Y_test,Y_pred1))
print("Recall is:",recall_score(Y_test,Y_pred1))
print("F1 score is:",f1_score(Y_test,Y_pred1))
```

# Random Forest Classifier

```python
#Implementing Random Forest Classifier on undersampling dataset

from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
#Make prediction

Y_pred2=rfc.predict(X_test)
```

```python
print("Accuracy score is:",accuracy_score(Y_test,Y_pred2))
print("Precision is:",precision_score(Y_test,Y_pred2))
print("Recall is:",recall_score(Y_test,Y_pred2))
print("F1 score is:",f1_score(Y_test,Y_pred2))
```
